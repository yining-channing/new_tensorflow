```python
import cv2
import os

# 视频文件的路径 
video_path = r'E:\shuju\video_origin\11_volvo.mp4'  # 替换为视频文件路径
output_folder = r'E:\shuju\video_frames'

# 创建保存帧的文件夹，如果不存在则创建
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
# 打开视频文件
cap = cv2.VideoCapture(video_path)

# 检查视频是否成功打开
if not cap.isOpened():
    print("Error: 无法打开视频文件")
else:
    print("视频成功打开")
```

```python

# 打开视频文件
cap = cv2.VideoCapture(video_path)

# 检查视频是否成功打开
if not cap.isOpened():
    print("Error: 无法打开视频文件")
else:
    # 获取视频总帧数
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    print(f"视频总帧数: {total_frames}")

    # 初始化帧计数器
    frame_count = 0

    # 设置每隔多少帧保存一次
    save_every_n_frames = 100  # 调整为每一帧保存一次

    # 读取视频帧
    while True:
        ret, frame = cap.read()

        # 如果没有更多帧，退出循环
        if not ret:
            print(f"在读取第 {frame_count} 帧时读取失败或结束，已退出循环。")
            break

        # 打印调试信息，确保每一帧都被正确读取
        print(f"Processing frame {frame_count}/{total_frames}")

        # 每隔多少帧保存一次
        if frame_count % save_every_n_frames == 0:
            # 保存帧为图片
            frame_filename = os.path.join(output_folder, f'frame_{frame_count:04d}.jpg')
            success = cv2.imwrite(frame_filename, frame)

            if success:
                print(f'Successfully saved: {frame_filename}')
            else:
                print(f'Failed to save: {frame_filename}')

        # 增加帧计数器
        frame_count += 1

    # 释放视频捕捉对象
    cap.release()

    print(f"总共处理了 {frame_count} 帧")

```

```python
# 列出保存的帧文件
extracted_frames = os.listdir(output_folder)
print(f"已保存的帧文件：{extracted_frames[:10]}")  # 仅显示前10个文件作为示例
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 加载并显示其中一个提取的帧 (这里显示第一个帧)
img_path = os.path.join(output_folder, extracted_frames[0])
img = mpimg.imread(img_path)
plt.imshow(img)
plt.axis('off')  # 隐藏坐标轴
plt.show()
```
